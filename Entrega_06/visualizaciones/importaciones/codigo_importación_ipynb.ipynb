```python
import pandas as pd
import altair as alt

# Cargar datos desde CSV
df = pd.read_csv("/content/pistacho grafico anual.csv", sep=";")

# Filtrar solo el producto 'pistacho'
df_pistacho = df[df['producto'].str.lower() == 'pistacho'].copy()

# Limpiar y convertir la columna 'Valor CIF acomulado' a número
df_pistacho['Valor CIF acomulado'] = df_pistacho['Valor CIF acomulado'] \
    .str.replace('.', '', regex=False) \
    .str.replace(',', '.', regex=False) \
    .astype(float)

# Ordenar por año
df_pistacho = df_pistacho.sort_values('año')

# Crear columna de color según el año (2025 naranja, otros verde)
df_pistacho['color'] = df_pistacho['año'].apply(lambda x: 'orange' if x == 2025 else 'green')

# Crear gráfico de barras con Altair
chart = alt.Chart(df_pistacho).mark_bar().encode(
    x=alt.X('año:O', title='Año', sort='ascending'),
    y=alt.Y('Valor CIF acomulado:Q', title='Valor CIF acumulado'),
    color=alt.Color('color:N', scale=None, legend=None)
).properties(
    width=600,
    height=400,
    title='Valor CIF acumulado del Pistacho por año (hasta diciembre, excepto 2025 que llega hasta septiembre)'
)

# Añadir etiquetas de valor encima de cada barra
text = chart.mark_text(
    align='center',
    baseline='bottom',
    dy=-5  # Desplazamiento vertical del texto
).encode(
    text=alt.Text('Valor CIF acomulado:Q', format=',.0f')
)

# Mostrar gráfico
chart + text
```

```python
import pandas as pd
import altair as alt

# --- 1. Cargar datos ---
df = pd.read_csv("/content/pistacho grafico por mes.csv", sep=";")

# --- 2. Limpieza de datos ---
df = df.dropna(axis=1, how="all")
df = df[["Producto", "Año", "Mes", "Valor en CIF"]]

df.columns = df.columns.str.strip()
df["Mes"] = df["Mes"].str.strip()
df["Valor en CIF"] = df["Valor en CIF"].astype(str).str.replace(",", ".")
df["Valor en CIF"] = pd.to_numeric(df["Valor en CIF"], errors="coerce")

df = df.dropna(subset=["Año", "Mes", "Valor en CIF"])
df = df[df["Valor en CIF"] > 0]

df["Año"] = df["Año"].astype(int)

# --- 3. Ordenar los meses ---
orden_meses = [
    "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
    "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
]

df["Mes"] = pd.Categorical(df["Mes"], categories=orden_meses, ordered=True)
df = df.dropna(subset=["Mes"])

# --- 4. Crear gráfico con paleta personalizada ---
custom_palette = ["#d1d983", "#7A9046", "#8B6B4A"]

chart = (
    alt.Chart(df)
    .mark_bar()
    .encode(
        x=alt.X("Año:O", title="Año"),
        y=alt.Y("sum(Valor en CIF):Q", title="Importaciones (USD)"),
        color=alt.Color(
            "Año:N",
            title="Año",
            scale=alt.Scale(range=custom_palette)
        ),
        column=alt.Column("Mes:N", title="Mes", sort=orden_meses, spacing=10),
        tooltip=[
            alt.Tooltip("Año:O"),
            alt.Tooltip("Mes:N"),
            alt.Tooltip("sum(Valor en CIF):Q", title="Importaciones (USD)", format=",")
        ]
    )
    .properties(
        title="Importaciones de Pistacho por mes y año (en USD)",
        width=100,
        height=120
    )
)

chart
```

```python
# --- Librerías ---
import pandas as pd
import altair as alt

# --- 1. Cargar el CSV ---
df = pd.read_csv("/content/almendras grafico por mes.csv", sep=";")

# --- 2. Limpieza de datos ---
df = df.dropna(axis=1, how="all")
df = df[["Producto", "Año", "Mes", "Valor en CIF"]]

df.columns = df.columns.str.strip()
df["Mes"] = df["Mes"].str.strip()
df["Valor en CIF"] = df["Valor en CIF"].astype(str).str.replace(",", ".")
df["Valor en CIF"] = pd.to_numeric(df["Valor en CIF"], errors="coerce")

df = df.dropna(subset=["Año", "Mes", "Valor en CIF"])
df = df[df["Valor en CIF"] > 0]

df["Año"] = df["Año"].astype(int)

# --- 3. Ordenar los meses ---
orden_meses = [
    "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
    "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
]

df["Mes"] = pd.Categorical(df["Mes"], categories=orden_meses, ordered=True)
df = df.dropna(subset=["Mes"])

# --- 4. Paleta personalizada ---
custom_palette = ["#d1d983", "#7A9046", "#8B6B4A"]

# --- 5. Crear gráfico ---
chart = (
    alt.Chart(df)
    .mark_bar()
    .encode(
        x=alt.X("Año:O", title="Año"),
        y=alt.Y("sum(Valor en CIF):Q", title="Importaciones (USD)"),
        color=alt.Color(
            "Año:N",
            title="Año",
            scale=alt.Scale(range=custom_palette)
        ),
        column=alt.Column("Mes:N", title="Mes", sort=orden_meses, spacing=10),
        tooltip=[
            alt.Tooltip("Año:O"),
            alt.Tooltip("Mes:N"),
            alt.Tooltip("sum(Valor en CIF):Q", title="Importaciones (USD)", format=",")
        ]
    )
    .properties(
        title="Importaciones de Almendras por mes y año (en USD)",
        width=100,
        height=120
    )
)

chart
```
