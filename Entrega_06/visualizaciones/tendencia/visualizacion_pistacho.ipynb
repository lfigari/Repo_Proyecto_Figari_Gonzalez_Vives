
# Visualización Google Trends — Pistachos, nueces y chocolate dubai (Chile)




```python

import pandas as pd
import matplotlib.pyplot as plt

csv_path = "/mnt/data/base_antigua (1).csv"

# Cargar datos: la exportación de Google Trends incluye dos filas de encabezado descriptivo,
# por lo que se omiten con `skiprows=2`.
df = pd.read_csv(csv_path, skiprows=2)

# Conversión de fechas
df['Semana'] = pd.to_datetime(df['Semana'])
df.head()
```

```python

plt.figure(figsize=(14,6))

color_pistachos = "#7A9046"  # verde pistacho
color_nueces = "#8B6B4A"        # café
color_chocdubai = "#d1d983"  # verde claro

plt.plot(df['Semana'], df['pistachos: (Chile)'], label='pistachos: (Chile)', color=color_pistachos, linewidth=2)
plt.plot(df['Semana'], df['nueces: (Chile)'], label='nueces: (Chile)', color=color_nueces, linewidth=2)
plt.plot(df['Semana'], df['chocolate dubai: (Chile)'], label='chocolate dubai: (Chile)', color=color_chocdubai, linewidth=2)

plt.title("Interés en el tiempo por término (Chile)", fontsize=18)
plt.xlabel("Semana", fontsize=12)
plt.ylabel("Índice", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.legend(frameon=False, fontsize=11)

plt.tight_layout()
plt.show()
```
